```python
import os
import pandas as pd
import numpy as np
import pyreadr
import config_91
import utils_91
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
fold = '1.0'
clf_cfg = config_91.classifier_config
HISTORY_DIR = clf_cfg['NN91_HISTORY_DIR']
with open(os.path.join(HISTORY_DIR, f'{fold}.pkl'), 'rb') as handle:
    fold_history = pickle.load(handle)

print(fold_history)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(fold_history['train_accs'])
plt.plot(fold_history['val_accs'])
plt.title(f'Fold {fold} accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc = 'best')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(fold_history['train_losses'])
plt.plot(fold_history['val_losses'])
plt.title(f'Fold {fold} loss')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc = 'best')
plt.show()
```
