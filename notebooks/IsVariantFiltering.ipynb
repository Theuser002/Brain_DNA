```python
ROOT = '/media/data/hungnt/work/SourceCode/Brain_DNA'
R_FEATURES_FOLDER = os.path.join(ROOT, 'data', 'raw', 'Features')
```

```python
fold = '1.0'
data = read_r(os.path.join(R_FEATURES_FOLDER, f'betas.{fold}.RData'))
train_df = data['betas.train']
test_df = data['betas.test']
train_cols = train_df.columns.values.tolist()
test_cols = test_df.columns.values.tolist()

for i in range(1, 6):
    for j in range(0, 6):
        temp_fold = f'{i}.{j}'
        temp_data = read_r(os.path.join(R_FEATURES_FOLDER, f'betas.{temp_fold}.RData'))
        temp_train_df = temp_data['betas.train']
        temp_test_df = temp_data['betas.test']
        temp_train_cols = sorted(temp_train_df.columns.values.tolist())
        temp_test_cols = sorted(temp_test_df.columns.values.tolist())
        print(f'{temp_fold}: {temp_train_cols == sorted(train_cols)} - {temp_test_cols == sorted(test_cols)} - {temp_train_cols == temp_test_cols}')
```
